# DATA.GOUV

```python
driver = webdriver.Chrome()
driver.get('https://recherche.data.gouv.fr/fr/faq')
```

```python
afficherbtn = driver.find_elements(By.XPATH,'/html/body/div[3]/main/div[3]/div/div[2]/div/button')[0]
afficherbtn.click()
```

## Reponses

```python
all_lists = []
for k in range(1, 21):
    current_list = []
    
    for i in range(1, 21):
        try:
            ans = driver.find_element(By.XPATH, f'/html/body/div[3]/main/div[3]/div/div[2]/ul[{k}]/li[{i}]/div[2]')
            current_list.append(ans.text)
        except :
            pass
    
    all_lists.append(current_list)

print(all_lists)

```

```python
flat_list = []
for sublist in all_lists:
    for item in sublist:
        flat_list.append(item)
print(flat_list)
```

```python
anss = flat_list
len(anss)
```

```python
import requests as req 
url = "https://deliveroo.fr/fr/faq"
response = req.get(url)
```

```python
import re 
html = response.text

notes = re.findall("<h3>(.*?)</h3>(.*?)(?=<h3>|$)",html)
notes
```

```python
reponses = [element[1:] for element in notes ]
print(reponses)
```

```python
len(reponses)
```

```python
liste_simple = []
for tuple_ in reponses:
    for element in tuple_:
        liste_simple.append(element)

print(liste_simple)
```
